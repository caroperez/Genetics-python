Reading of the BRCA1 gene, which encodes a tumor suppressor gene It is often found mutated in many types of cancers, especially breast cancer.

```python
# load fasta file
with open('fasta_seq_files/brca1_homo_sapiens_wt.fasta', 'r') as fasta_file:
    bcra1 = fasta_file.readlines()
```

```python
# segregate info and nucleotide sequence from data
info_seq = bcra1[0] # info from gene
bcra1_seq_raw = bcra1[1:] # nucleotide sequence data
```

```python
# breaking down gene information
info_total = info_seq.split(" ")

specie = ' '.join(info_total[1:3])
chromosome = ' '.join(info_total[3:5]).replace(',', '')
genome = genome = ''.join(info_total[5])

# print info
print('brca1, genome: {} specie: {} chromosome: {}'.format(genome, specie, chromosome))
```

```python
bcra1_seq = ''.join(bcra1_seq_raw).replace('\n', '') # remove jump lines and bind secuence line in one
```

```python
# functions
def gene_info(seq_fasta):
    '''Function for obtaining information about a gene through its fasta sequence'''
    info_seq = seq_fasta[0].split(' ')
    specie = ' '.join(info_total[1:3])
    chromosome = ' '.join(info_total[3:5]).replace(',', '')
    genome = ''.join(info_total[5])
    return specie, chromosome, genome

def gene_seq(seq_fasta):
    '''Función para obtener la secuencia de un gen a través de su secuencia fasta'''
    seq_gene_raw = seq_fasta[1:]
    seq_gene = ''.join(seq_gene_raw).replace('\n', '')
    return seq_gene

# All in one function
def gene_from_fasta(gene):
    '''function that returns the info, number of nucleotides 
    and sequence of a gene contained in a fasta file'''
    gene_info(gene)
    seq_gene = gene_seq(gene)
    print('Genome: {}, Specie: {}, Chromosome: {}, n.nucleotids: {}'.format(genome, specie, chromosome, len(seq_gene)))
    print('Sequence:')
    print(seq_gene)
```

```python
gene_from_fasta(bcra1)
```
